```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv', sep =';')
```

The top internet providers in Portugal are MEO, Vodafone, NOS, and NOWO

Vodafone : 0.75 euro per minute or 0.0125 per second

```python
df["biaya_per_campaign"] = df["duration"] * 0.0125
biaya_campaign = df["biaya_per_campaign"]
print(biaya_campaign)
```

```python
sns.histplot(df['biaya_per_campaign'], color='blue', bins=10)
plt.title('Biaya per campaign provider Vodafone')
plt.show()
```

```python
df['biaya_per_campaign'].plot(kind='hist', figsize=(10, 6), bins=50)
plt.title('biaya per campaign', size=16)
plt.ylabel('Number of data', size=14)
plt.xlabel('Prices', size=14)
plt.show()
```

Berdasarkan data, biaya maksimal yang dikeluarkan perusahaan untuk melakukan telepon campaign yakni sebesar 61.47 euro dan nilai minimalnya yakni 0 (tidak diangkat teleponnya). Sedangkan durasi rata-rata perusahaan melakukan panggilan telepon yakni selama 4 menit dengan mengeluarkan biaya sebesar 3.22 euro
